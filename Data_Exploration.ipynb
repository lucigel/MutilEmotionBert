```python
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline #tranh loi hien thi
```

```python
# read data
# let's use train data to understand this dataset
data = pd.read_csv('nlp_train.csv', index_col='id')
data.sample(5)
```

```python
# basic stats
data.describe()
```

> Don't have Missing Values

```python
data['anger'].value_counts()
```

```python
freqs = {"anger":data['anger'].value_counts()[1],
"anticipation": data['anticipation'].value_counts()[1],
"disgust":data['disgust'].value_counts()[1],
"fear":data['fear'].value_counts()[1],
"joy":data['joy'].value_counts()[1],
"love":data['love'].value_counts()[1],
"optimism":data['optimism'].value_counts()[1],
"pessimism":data['pessimism'].value_counts()[1],
"sadness":data['sadness'].value_counts()[1],
"surprise":data['surprise'].value_counts()[1],
"trust":data['trust'].value_counts()[1],
"neutral":data['neutral'].value_counts()[1]}
```

```python
freqs 
```

## Use matplotlib to visualize class of distribution

```python
plt.figure(figsize=(13,5))
plt.bar(freqs.keys(), freqs.values(), color = 'skyblue', width=0.5)
plt.title("Class of Distribution About Emotion in Text")
x = plt.xlabel("Emotion", fontsize = 13)
x.set_color('r')
y = plt.ylabel('Amount of Each class', fontsize = 13)
y.set_color('r')
plt.savefig('Distribution of class')
plt.show()
```
